# 3) Multivariate Analysis

Multivariate yaitu menggabungkan beberapa kolom dengan tujuan menemukan hubungan antara satu kolom dengan kolom lainnya.

- Berapakah jumlah pelanggan berpotensi pindah berlangganan berdasarkan gender?

```python
#Memvisualisasikan data menggunakan countplot
sns.countplot(data=df,x="gender",hue="Churn")
```

Setelah dibuat visualisasi data, Pelanggan wanita&pria yang berpotensi pindah berlangganan lebih sedikit jumlahnya dibandingkan yang tidak pindah berlangganan.

- Berapa lama waktu pelanggan berlangganan merasakan ingin pindah perusahaan untuk berlangganan? 

```python
sns.boxplot(data=df,x="tenure",y="Churn")
```

Hasil visualisasi menunjukkan bahwa customer yang ingin pindah perusahaan berlangganan(YES) yaitu customer yang sudah berlanggganan selama rentang waktu 1-30 bulan, dan boxplot churn YES juga menghasilkan data miring(tidak simetris) ditunjukkan oleh boxplot dengan letak garis median tidak berada di tengah kotak. Data miring juga ditandai dari panjang whisker tidak sama antara kanan-kiri. Whisker kanan lebih panjang dan terdapat outlier di bagian kanan yang menandakan data cenderung miring ke kanan (positive skewness).

Sedangkan customer yang tidak ingin pindah perusahaan berlangganan(NO) yaitu customer yang sudah berlanggganan selama rentang waktu 15-60 bulan, boxplot churn NO merupakan data simetris(berdistribusi normal) ditunjukkan oleh boxplot dengan garis median berada di tengah kotak. Selain itu data simetris juga ditunjukkan oleh panjang whisker kiri dengan panjang whisker kanan, serta tidak terdapat nilai outlier.

- Membuat visualisasi data menggunakan horizontal boxplot berdasarkan hubungan label "MonthlyCharger" dan "Churn".

Kemudian melakukan proses dummy variable traps untuk 
mengkuantitatifkan variable yang bersifat kualitatif(tipe data object)

Contoh: gender, seniorcitizen,partner,dependents, dll

```python
#get dummies to transform data 
df2=pd.get_dummies(df,drop_first=True)
df2.info()
```

- mengecek korelasi variabel "churn_yes" bertujuan untuk melihat fitur mana yang berhubungan kuat

```python
#correlation

# 0 no correlation
# 1 positive correlation
# -1 negative correlation

df3=df2.corr()['Churn_Yes'].sort_values().reset_index()

df3
```

hasil korelasi diatas menunjukkan bahwa ada beberapa label yang bernilai positif, yang dimana artinya label tersebut berpengaruh dengan label churn_yes. antara lain:
1. InternetService_Fiber optic
2. PaymentMethod_Electronic check	
3. MonthlyCharges

- syntax dibawah berfungsi untuk membuat variabel baru yaitu df3 dengan menghilangkan label/fitur "churn_yes"

```python
#drop Churn_Yes
df3=df3[~df3['index'].str.contains('Churn_Yes')]
```

- Kemudian membuat visualisasi data berupa barplot untuk melihat korelasi antar label

```python
chart = sns.barplot(data=df3,x="index",y="Churn_Yes")

for item in chart.get_xticklabels():
  item.set_rotation(90)
```

Hasil visualisasi diatas menunjukkan korelasi atribut churn yes terhadap atribut lain bernilai yang kuat(bernilai positif) antara lain:
1. InternetService_Fiber optic
2. PaymentMethod_Electronic check	
3. MonthlyCharges

- membuat visualisasi data mengunakan heatmap untuk mengecek korelasi antar label

```python
#Heatmap to check correlation
sns.heatmap(df2.corr(),cmap="Blues")
```
